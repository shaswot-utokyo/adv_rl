```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time = np.linspace(0,24,100)
g_90 = [0.90**i for i in time ]
g_95 = [0.95**i for i in time]
g_97 = [0.97**i for i in time]
g_98 = [0.98**i for i in time]
g_99 = [0.99**i for i in time]
```

```python
fig = plt.figure()
plt.plot(time, g_90, label='g90')
plt.plot(time, g_95, label='g95')
plt.plot(time, g_97, label='g97')
plt.plot(time, g_98, label='g98')
plt.plot(time, g_99, label='g99')
plt.grid()
fig.legend()
```
